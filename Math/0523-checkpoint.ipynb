```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
rv = stats.norm() #정규분포 제작
xs = np.linspace(-4,4,100)
ys = rv.pdf(xs)
plt.plot(xs,ys)
plt.show()
```

```python
# 히스토그램 그리기
rv = stats.norm() 
xs = np.linspace(-4,4,100)
ys = rv.pdf(xs)
sample_size = int(1e6)
n_sample = rv.rvs(sample_size)
plt.hist(n_sample, bins=100, density=True) #density 그래프 범위 압축해서 보여줌

plt.plot(xs,ys)
plt.show()
```

```python
rv = stats.norm(2,0.5) 
xs = np.linspace(0,4,1000) # 0-4까지 1000개로 나눔
ys = rv.pdf(xs)
ys1 = rv.cdf(xs)

sample_size = int(1e6)
n_sample = rv.rvs(sample_size)
plt.hist(n_sample, bins=100, density=True) 

plt.plot(xs,ys, ls='-', label='f(x)')
plt.plot(xs,ys1, ls='--', label='F(x)')
plt.legend()
plt.show()
```
