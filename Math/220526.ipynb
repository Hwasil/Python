```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n=10
rv=stats.norm()
rv1=stats.norm()
xs=np.linspace(0, 30, 100)

ys=rv1.pdf(xs)
sample_size=int(1e6)

tmp_sample=rv.rvs(n, sample_size)
chi2_sample=np.sum(tmp_sample**2, axis=0)
plt.hist(chi2_sample, bin=100, density=True, range=(0, 30), label="chi2(10)")

plt.legend()
plt.show()
```

```python
#for n, ls in zip([3, 5, 10], ['-', '--', ':']) #구현
rv1=stats.chi2(3)
rv2=stats.chi2(5)
rv3=stats.chi2(10)

xs=np.linspace(0, 20, 1000)
plt.plot(xs,  rv1.pdf(xs), label='chi(3)', ls='-')
plt.plot(xs,  rv2.pdf(xs), label='chi(5)', ls='--')
plt.plot(xs,  rv3.pdf(xs), label='chi(10)', ls=':')

plt.legend()
plt.show()
```
