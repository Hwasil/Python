```python
#Chapter 08
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
n=10
rv=stats.norm()
rv1=stats.chi2(10)

sample_size=int(1e6)
z_sample=rv.rvs(sample_size)
ch_sample=rv1.rvs(sample_size)
t_sample=z_sample/np.sqrt(ch_sample/n)

plt.hist(t_sample, bins=100, density=True)
```

```python
# p.226-227
rv3=stats.norm(0, 1)
xs=np.linspace(-3, 3, 1000)
for n,ls in zip([3, 5, 10], ['-', '--', ':']):
    rv=stats.t(n)
    plt.plot(xs, rv.pdf(xs), ls=ls, label=f't({n})')

plt.plot(xs, rv3.pdf(xs), ls='-.', label='N(0,1)')

plt.legend()
plt.show()
```

```python
# p.231-232 
xs=np.linspace(0, 6, 200)[1:]
for n,ls in zip([3, 5, 10], ['-', '--', ':']):
    rv=stats.f(n, 10)
    plt.plot(xs, rv.pdf(xs), ls=ls, label=f'F({n}, 10)')

plt.legend()
plt.show()
```
